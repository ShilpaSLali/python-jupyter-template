### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

def validate_data(data):
    if not all(isinstance(i, (int, float)) for i in data):
        raise TypeError("All data values must be numeric.")

# Define the data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

try:
    validate_data(data)

    # Create the histogram
    plt.figure(figsize=(6, 4))
    plt.hist(data, bins=5, color='purple', edgecolor='black', alpha=0.7)

    # Customize the plot
    plt.title('Histogram of Data')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    # Show the plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
