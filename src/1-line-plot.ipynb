### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

def validate_data(x, y):
    if len(x) != len(y):
        raise ValueError("X and Y values must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise TypeError("X and Y values must be numeric.")

# Generate x values
x = np.linspace(0, 10, 100)
y = x ** 2  # Compute y values as the square of x

try:
    validate_data(x.tolist(), y.tolist())

    # Create the plot
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='y = x²', color='b', linewidth=2)

    # Customize the plot
    plt.title('Square Function Plot')
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()

    # Show the plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
