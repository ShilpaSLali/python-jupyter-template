### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data validation functions
def validate_data(x, y):
    if len(x) != len(y):
        raise ValueError("X and Y values must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise TypeError("X and Y values must be numeric.")

# Data
x = np.linspace(0, 10, 100)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

try:
    validate_data(x.tolist(), y.tolist())
    validate_data(x_scatter, y_scatter)

    # Line Chart
    axes[0, 0].plot(x, y, color='b', linewidth=2)
    axes[0, 0].set_title('Line Chart')

    # Bar Chart
    axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axes[0, 1].set_title('Bar Chart')

    # Scatter Plot
    axes[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
    axes[1, 0].set_title('Scatter Plot')

    # Pie Chart
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)
    axes[1, 1].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
