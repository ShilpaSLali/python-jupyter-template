### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

def validate_data(labels, sizes):
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must have the same length.")
    if sum(sizes) == 0:
        raise ValueError("The sum of sizes must be greater than zero.")
    if not all(isinstance(size, (int, float)) and size >= 0 for size in sizes):
        raise TypeError("Sizes must be non-negative numeric values.")

# Define data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

try:
    validate_data(labels, sizes)

    # Create the pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)

    # Set title
    plt.title('Programming Language Usage Distribution')

    # Show the plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
